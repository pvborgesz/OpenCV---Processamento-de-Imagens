```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv 
```

```python
img = cv.imread('images/letter-p-smaller.png', cv.COLOR_RGB2BGR) #Lendo imagem
img2 = cv.imread('images/letter-b.jpg', 1) #Lendo imagem

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img2),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
kernel = np.ones((5,5),np.uint8)

dilation = cv.dilate(img,kernel,iterations = 3)
img2_dilation = cv.dilate(img2,kernel,iterations = 10)
# plt.subplot(122),plt.imshow(dilation),plt.title('Após Dilatação')
# plt.xticks([]), plt.yticks([])
# plt.show()

plt.subplot(121),plt.imshow(dilation),plt.title('Após Dilatação')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img2_dilation),plt.title('Após Dilatação')
plt.xticks([]), plt.yticks([])
plt.show()

```

```python
erosion = cv.erode(img,kernel,iterations = 2) # Fazendo erosão
img2_erosion = cv.erode(img2,kernel,iterations = 20)

plt.subplot(121),plt.imshow(erosion),plt.title('Após Erosão')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img2_erosion),plt.title('Após Erosão')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cv.subtract

img_diff = cv.absdiff(erosion, dilation)
img2_diff  = cv.absdiff(img2_erosion, img2_dilation)
# plt.subplot(122),plt.imshow(img2),plt.title('Após Subtração da Dilatação Pela Erosão')
# plt.xticks([]), plt.yticks([])
# plt.show()

plt.subplot(121),plt.imshow(img_diff),plt.title('Subtração da Dilatação Pela Erosão\n')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img2_diff),plt.title('Após Subtração da Dilatação Pela Erosão')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img3 = cv.bitwise_or(img_diff,dilation,img_diff)
img4 = cv.morphologyEx(img2, cv.MORPH_GRADIENT, kernel) # Morphological Gradient
img4 = cv.bitwise_or(img4,img2_dilation,img4)
# A given pixel is turned “off” if it has a value of zero, 
# and it is turned “on” if the pixel has a value greater than zero.
#Operação OR bit a bit na imagem:
#Disjunção bit a bit dos elementos da matriz de entrada.

plt.subplot(121),plt.imshow(img3),plt.title('Após bitwise or')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img4),plt.title('Após bitwise or')
plt.xticks([]), plt.yticks([])
plt.show()
```
